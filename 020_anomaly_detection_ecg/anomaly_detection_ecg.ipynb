# Autoencoder

```python
import os
os.environ["CUDA_VISIBLE_DEVICES"] = "-1"


import pandas
URL = "http://storage.googleapis.com/download.tensorflow.org/data/ecg.csv"
t = pandas.read_csv(URL, header=None)

x = t.iloc[:,0:-1]
y = t.iloc[:,-1]

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1)

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(x_train)
x_train_scaled = scaler.transform(x_train)
x_test_scaled  = scaler.transform(x_test)

# Mit x_train_normal (und nur hiermit) werden wir den Anomalie-Detektor trainieren!
x_train_normal   = x_train_scaled[ y_train.astype(bool) ]
x_train_abnormal = x_train_scaled[ ~y_train.astype(bool) ]

# Mit diesen werden wir den Anomalie-Detektor testen
x_test_normal   = x_test_scaled[ y_test.astype(bool) ]
x_test_abnormal = x_test_scaled[ ~y_test.astype(bool) ]


from tensorflow.keras.models import Model, Sequential
from tensorflow.keras import layers, losses

class AnomalyDetector(Model):

    def __init__(self):
        super(AnomalyDetector, self).__init__()
        self.encoder = Sequential([
            layers.Dense(32, activation="relu"),
            layers.Dense(16, activation="relu"),
            layers.Dense(8, activation="relu")
        ])

        self.decoder = Sequential([
            layers.Dense(16, activation="relu"),
            layers.Dense(32, activation="relu"),
            layers.Dense(140, activation="sigmoid")
        ])


    def call(self, x):

        encoded = self.encoder(x)
        decoded = self.decoder(encoded)
        return decoded

model = AnomalyDetector()
model.compile(optimizer="adam", loss="mae")
h = model.fit(x_train_normal, x_train_normal, epochs=100, verbose=0)

import matplotlib.pyplot as plt
plt.plot( h.history["loss"] )
plt.xlabel("Epoche")
plt.ylabel("MAE Loss")
plt.show()        
```

# Den Anomalieklassifikator bauen

```python
def classify(autoenc, reco_error_threshold, timeseries):
    
    reco_timeseries = autoenc.predict( timeseries.reshape(1,140), verbose=0 ).flatten()    
    reco_error = mean_absolute_error(timeseries, reco_timeseries)    
    if reco_error < reco_error_threshold:
        return 1 # normal time series
    else:
        return 0 # abnormal time series


timeseries = x_test_normal[3]
classification_result = classify(model, reco_error_threshold, timeseries)
if classification_result == 1:
    print("Normales ECG!")
else:
    print("Achtung! Abnormales ECG!")
```

# Alles in einem Skript

```python
import os
os.environ["CUDA_VISIBLE_DEVICES"] = "-1"


# 1. Datensatz mit Zeitreihen holen
#    Zeitreihe hat 1 Feature
URL = "http://storage.googleapis.com/download.tensorflow.org/data/ecg.csv"
import pandas
t = pandas.read_csv(URL, header=None)

# 2. Tabelle in Zeitreihen x + Label y (1=normal, 0=abnormal) splitten
x = t.iloc[:,0:-1]
y = t.iloc[:,-1]

# 3. Daten fÃ¼rs ML vorbereiten
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1)

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(x_train)
x_train_scaled = scaler.transform(x_train)
x_test_scaled  = scaler.transform(x_test)

# Mit x_train_normal (und nur hiermit) werden wir den Anomalie-Detektor trainieren!
x_train_normal   = x_train_scaled[ y_train.astype(bool) ]
x_train_abnormal = x_train_scaled[ ~y_train.astype(bool) ]

# Mit diesen werden wir den Anomalie-Detektor testen
x_test_normal   = x_test_scaled[ y_test.astype(bool) ]
x_test_abnormal = x_test_scaled[ ~y_test.astype(bool) ]


# 4. Autoencoder auf normale ECGs "einschleifen"
from tensorflow.keras.models import Model, Sequential
from tensorflow.keras import layers, losses

class AnomalyDetector(Model):

    def __init__(self):
        super(AnomalyDetector, self).__init__()
        self.encoder = Sequential([
            layers.Dense(32, activation="relu"),
            layers.Dense(16, activation="relu"),
            layers.Dense(8, activation="relu")
        ])

        self.decoder = Sequential([
            layers.Dense(16, activation="relu"),
            layers.Dense(32, activation="relu"),
            layers.Dense(140, activation="sigmoid")
        ])


    def call(self, x):

        encoded = self.encoder(x)
        decoded = self.decoder(encoded)
        return decoded

model = AnomalyDetector()
model.compile(optimizer="adam", loss="mae")
h = model.fit(x_train_normal, x_train_normal, epochs=100, verbose=0)

import matplotlib.pyplot as plt
plt.plot( h.history["loss"] )
plt.xlabel("Epoche")
plt.ylabel("MAE Loss")
plt.show()        

# 5. Reko-Fehler-Schwelle bestimmen
x_train_normal_reconstructions = model.predict(x_train_normal)
from sklearn.metrics import mean_absolute_error
errors = mean_absolute_error(x_train_normal.T,
                             x_train_normal_reconstructions.T,
                             multioutput="raw_values")
reco_error_threshold = errors.mean() + errors.std()

# 6. Klassifikatorkopf erstellen
def classify(autoenc, reco_error_threshold, timeseries):
    
    reco_timeseries = autoenc.predict( timeseries.reshape(1,140), verbose=0 ).flatten()    
    reco_error = mean_absolute_error(timeseries, reco_timeseries)    
    if reco_error < reco_error_threshold:
        return 1 # normal time series
    else:
        return 0 # abnormal time series


timeseries = x_test_normal[0]
classification_result = classify(model, reco_error_threshold, timeseries)
if classification_result == 1:
    print("Normales ECG!")
else:
    print("Achtung! Abnormales ECG!")
```

```python
correct = 0
N = x_test_normal.shape[0]
for i in range(N):
    timeseries = x_test_normal[i]
    classification_result = classify(model, reco_error_threshold, timeseries)
    if classification_result == 1:
        correct += 1        
    
print(f"In {correct} von {N} FÃ¤llen hat der Anomaliedetektor korrekt klassifiziert!")
```
