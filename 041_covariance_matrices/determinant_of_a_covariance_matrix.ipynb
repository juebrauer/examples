# What does a determinant do?

The determinant of a covariance matrix measures how much the data points are spread out overall.

```python
import numpy

mu = [5.0, 5.0]

C = [[1.0,  0.0],
     [0.0,  10.0]]

C = [[5.0,  0.0],
     [0.0,  10.0]]

#C = [[25.0,  0.0],
#     [0.0,  10.0]]

D = numpy.random.multivariate_normal(mean=mu, cov=C, size=1000)
determinant = numpy.linalg.det( C )


import matplotlib.pyplot as plt
plt.scatter(D[:,0], D[:,1])
plt.axis("equal")
plt.xlim(-30,+30)
plt.ylim(-30,+30)
plt.title(f"Determinant is: {determinant:.2f}")
plt.grid()
plt.show()
```

# Calculating the Determinant

## 2×2 Matrix

For

$$
A =
\begin{pmatrix}
a & b \\
c & d
\end{pmatrix}
$$

we have:

$$
\det(A)=ad-bc
$$

Example:

$$
A =
\begin{pmatrix}
1 & 2 \\
3 & 4
\end{pmatrix}
$$

$$
\det(A)=1\cdot 4 - 2\cdot 3 = 4-6=-2
$$

---

## 3×3 Matrix

For

$$
A =
\begin{pmatrix}
a & b & c \\
d & e & f \\
g & h & i
\end{pmatrix}
$$

we have:

$$
\det(A)
=
aei+bfg+cdh-ceg-bdi-afh
$$

Equivalently:

$$
\det(A)
=
a(ei-fh)-b(di-fg)+c(dh-eg)
$$

Example:

$$
A =
\begin{pmatrix}
1 & 2 & 3 \\
0 & 1 & 4 \\
5 & 6 & 0
\end{pmatrix}
$$

$$
\det(A)
=
1(1\cdot0-4\cdot6)
-2(0\cdot0-4\cdot5)
+3(0\cdot6-1\cdot5)
$$

$$
= -24 + 40 - 15 = 1
$$

## n×n Matrix

For a matrix

$$
A=(a_{ij}) \in \mathbb{R}^{n\times n}
$$

the determinant can be defined using the Leibniz formula:

$$
\det(A)
=
\sum_{\sigma \in S_n}
\operatorname{sgn}(\sigma)
\prod_{i=1}^{n} a_{i,\sigma(i)}
$$

Where:

- $S_n$: the set of all permutations of $\{1,\dots,n\}$
- $\operatorname{sgn}(\sigma)$: the sign of the permutation
- $\prod_{i=1}^{n} a_{i,\sigma(i)}$: the product of selected matrix entries

In practice, large determinants are usually computed using Gaussian elimination:

$$
A \rightarrow U
$$

where $U$ is an upper triangular matrix. Then:

$$
\det(U)=u_{11}u_{22}\cdots u_{nn}
$$

So the determinant is the product of the diagonal entries.

Important rules for row operations:

- Swapping two rows changes the sign of the determinant.
- Multiplying a row by $k$ multiplies the determinant by $k$.
- Adding a multiple of one row to another row does not change the determinant.
