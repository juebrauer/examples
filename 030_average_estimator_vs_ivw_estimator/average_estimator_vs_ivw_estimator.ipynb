```python

import numpy as np
import matplotlib.pyplot as plt

def get_mae(sensor_data, true_value):
    error = np.mean(np.abs(sensor_data - true_value))
    error_str = f"{error:.2f}"
    return error_str
    
#np.random.seed(42) # für reproduzierbare Ergebnisse

true_distance = 100.0 # tatsächliche Entfernung zum Objekt

# Standard deviations of the noise for each sensor
s1_noise = 2.0 # sehr präziser Sensor
s2_noise = 5.0 # mittlerere Präzision
s3_noise = 10.0 # unpräziser Sensor

num_measurements = 100
s1_data = true_distance + np.random.normal(0, s1_noise, num_measurements)
s2_data = true_distance + np.random.normal(0, s2_noise, num_measurements)
s3_data = true_distance + np.random.normal(0, s3_noise, num_measurements)

mean_fusion = (s1_data + s2_data + s3_data) / 3

sum_of_ivw_weights = (1/s1_noise**2 + 1/s2_noise**2 + 1/s3_noise**2)
ivw_fusion = (s1_data / s1_noise**2 + s2_data / s2_noise**2 + s3_data / s3_noise**2) / sum_of_ivw_weights

error_s1 = get_mae(s1_data, true_distance)
error_s2 = get_mae(s2_data, true_distance)
error_s3 = get_mae(s3_data, true_distance)
error_fusion = get_mae(mean_fusion, true_distance)

alpha_sensors = 0.3
plt.plot(s1_data, label=f'Sensor 1 (präzise) MAE: {error_s1}', alpha=alpha_sensors)
plt.plot(s2_data, label=f'Sensor 2 (mittlere Präzision) MAE: {error_s2}', alpha=alpha_sensors)
plt.plot(s3_data, label=f'Sensor 3 (unpräzise) MAE: {error_s3}', alpha=alpha_sensors)
plt.plot(mean_fusion, label=f'Mittelwert-Fusion MAE: {error_fusion}', color='black', linewidth=2)
plt.plot(ivw_fusion, label=f'IVW-Fusion MAE: {get_mae(ivw_fusion, true_distance)}', color='red', linewidth=2)
plt.axhline(true_distance, color='gray', linestyle='--', label='Tatsächliche Entfernung')
plt.legend()
plt.show()
```
