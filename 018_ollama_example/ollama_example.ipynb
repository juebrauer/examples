```python
txt1 = """Ich habe die Waschmaschine bei Ihnen im Januar 25 gekauft.
Jetzt läuft sie ständig aus. Ich will ne neue!
"""

txt2 = """Es gibt doch diese neue Spülmaschine Bosch X5000. Geht die auch statt
mit Spültabs mit diesen umweltfreundlichen Spültüchern?"""

txt3 = """Ich habe Zahnschmerzen. Was kann ich da machen?"""

prompt = f"""Du sollst Kundenanfragen klassifizieren. Klassifiziere die folgende Anfrage:
---
{txt1}
---
Als Klassifikationsergebnis bitte nur eine Zahl ausgeben:
1.) Adressänderung
2.) Garantieanspruch
3.) Anfrage nach Produktinformationen
4.) Sonstiges
"""

print(prompt)
```

```python
from ollama import chat
from ollama import ChatResponse

response: ChatResponse = chat(model='gpt-oss:20b', messages=[
  {
    'role': 'user',
    'content': prompt,
  },
])
#print(response['message']['content'])
# or access fields directly from the response object
print(response.message.content)
```
